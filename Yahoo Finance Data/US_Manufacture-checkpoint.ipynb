### Import Dependencies 

```python
import yfinance as yf 
import pandas as pd 
import datetime
```

```python
# Create list of all companies we want records for. 
US_Stock_list = ['GM', 'F', 'FCAU', 'TSLA', 'LNC' ]

# Set Range we want to records from. 
start = datetime.date(2019,12,2) 
end = datetime.date(2020,12,3)


# Loop through records for all ticker symbols in our list.
for stock in US_Stock_list:
     globals()[stock] = yf.download(stock,start,end)
```

```python
# Add Company columns to all data frames for merge.
GM['Company'] = 'General Motors'
F['Company'] = 'Ford'
FCAU['Company'] = 'Chrysler'
TSLA['Company'] = 'Tesla'
LNC['Company'] = 'Lincoln'


df_list = [GM, F, FCAU, TSLA, LNC]

us_df = pd.concat(df_list)

us_df_reordered = ['Company', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']

final_us_df = us_df[us_df_reordered]
final_us_df
```

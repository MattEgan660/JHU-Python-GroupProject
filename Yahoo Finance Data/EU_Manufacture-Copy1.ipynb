### Import Dependencies 

```python
import yfinance as yf 
import pandas as pd 
import datetime
```

```python
# Create list of all companies we want records for. 
EU_Stock_list = ['VWAGY', 'BAMXF', 'DMLRY', 'POAHY', 'VLVLY' ]

# Set Range we want to records from. 
start = datetime.date(2019,12,2) 
end = datetime.date(2020,12,3)


# Loop through records for all ticker symbols in our list.
for stock in EU_Stock_list:
     globals()[stock] = yf.download(stock,start,end)
```

```python
# Add Company columns to all data frames for merge.
VWAGY['Company'] = 'Volkswagen'
BAMXF['Company'] = 'BMW'
DMLRY['Company'] = 'Mercedes'
POAHY['Company'] = 'Porshe'
VLVLY['Company'] = 'Volvo'


df_list = [VWAGY, BAMXF, DMLRY, POAHY, VLVLY]

eu_df = pd.concat(df_list)

eu_df_reordered = ['Company', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']

final_eu_df = eu_df[eu_df_reordered]
final_eu_df
```

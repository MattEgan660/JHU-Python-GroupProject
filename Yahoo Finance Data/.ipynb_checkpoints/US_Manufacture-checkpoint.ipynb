### Import Dependencies 

```python
import yfinance as yf 
import pandas as pd 
from pandas import Series,DataFrame
import numpy as np

import datetime
#Visualisation imports
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

#define csv output
output_data = ('../df_csv/us_df.csv')
```

```python
# Create list of all companies we want records for. 
US_Stock_list = ['GM', 'F', 'FCAU', 'TSLA' ]

# Set Range we want to records from. 
start = datetime.date(2019,12,2) 
end = datetime.date(2020,12,3)


# Loop through records for all ticker symbols in our list.
for stock in US_Stock_list:
     globals()[stock] = yf.download(stock,start,end)
```

```python
# Add Company columns to all data frames for merge.
GM['Company'] = 'General Motors'
F['Company'] = 'Ford'
FCAU['Company'] = 'Chrysler'
TSLA['Company'] = 'Tesla'


# Create list of manufacturer dataframes to create single dataframe. 
df_list = [GM, F, FCAU, TSLA]

# Create single dataframe with all U.S. manufacturers.
us_df = pd.concat(df_list)

# Add region column to dataframe.
us_df['Region'] = 'U.S.'

# Order Columns
us_df_reordered = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume', 'Company', 'Region']

# Display final dataframe
final_us_df = us_df[us_df_reordered]
final_us_df
```

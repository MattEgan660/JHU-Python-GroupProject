```python
# Import Required Modules
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import csv 
import glob

```

```python
# Create Dataframe with all manufacturers

path = './Car_data/'
all_files = glob.glob(path + "/*.csv")

file_list = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    file_list.append(df)

all_kbb = pd.concat(file_list, axis=0, ignore_index=True)
```

```python
all_kbb
```

```python
all_kk_groupby=all_kbb.groupby('Company')['MSRP'].median().reset_index()
print(all_kk_groupby)
```

```python
all_kk_max=all_kbb.groupby('Company')['MSRP'].max().reset_index()
print(all_kk_max)
```

```python
all_kk_min=all_kbb.groupby('Company')['MSRP'].min().reset_index()
print(all_kk_min)
```

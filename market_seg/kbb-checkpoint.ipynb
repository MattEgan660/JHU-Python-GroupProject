```python
# Import Required Modules
import pandas as pd 
import matplotlib as plt
import numpy as np
import seaborn as sns
import datetime as dt
import yfinance as yf
import csv 
import glob

#create csv of main df for other analyses
output_main_df = ('./df_csv/all_manuf.csv')
```

```python
# Create Dataframe with all manufacturers

path = './market_seg/Car_data/'
all_files = glob.glob(path + "/*.csv")

file_list = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    file_list.append(df)

all_kbb = pd.concat(file_list, axis=0, ignore_index=True)
all_kbb.to_csv(output_main_df)
```
